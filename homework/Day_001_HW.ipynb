#### 請寫一個函式用來計算 Mean Square Error  ( 用程式碼打出公式好酷)

$ MSE = \frac{1}{n}\sum_{i=1}^{n}{(Y_i - \hat{Y}_i)^2} $  

### Hint: [如何取平方](https://googoodesign.gitbooks.io/-ezpython/unit-1.html)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_squared_error(y,y_hat):
    MSE=1/len(y) * sum((y-y_hat)**2)
    return(MSE)

def mean_absolute_error(y, y_hat):
    MAE=1/len(y) * sum(abs(y-y_hat))
    return(MAE)
```

```python
w = 3
b = 0.5

x_lin = np.linspace(0, 100, 101) # np.linspace? --> np.linespace(start,end,num) 0~100共有101個整數

y = (x_lin + np.random.randn(101) * 5) * w + b  # np.random.randn(101) EQUAL TO  rnorm(101) in R

plt.plot(x_lin, y, 'b.', label = 'data points')  # 'b.' means Blue Point  [COLOR&MARKERS](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
y_hat = x_lin * w + b
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# 執行 Function, 確認有沒有正常執行
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
