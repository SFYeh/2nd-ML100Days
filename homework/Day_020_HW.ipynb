```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/house_train.csv/'
df_train = pd.read_csv(data_path + 'train.csv')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
train_Y 
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df['1stFlrSF'].describe()
```

```python
upper=df['1stFlrSF'].describe()['75%']+1.5*(df['1stFlrSF'].describe()['75%']-df['1stFlrSF'].describe()['25%'])
lower=df['1stFlrSF'].describe()['25%']-1.5*(df['1stFlrSF'].describe()['75%']-df['1stFlrSF'].describe()['25%'])
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
df['1stFlrSF'] = df['1stFlrSF'].clip(lower,upper)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

<span style="color:blue">以本題來說，捨棄離群值之表現優於調整離群值</span>

```python
df
```

```python
df2=df.copy()
train_Y2=train_Y.copy()
```

```python
upper=df['1stFlrSF'].describe()['75%']+1.5*(df['1stFlrSF'].describe()['75%']-df['1stFlrSF'].describe()['25%'])
lower=df['1stFlrSF'].describe()['25%']-1.5*(df['1stFlrSF'].describe()['75%']-df['1stFlrSF'].describe()['25%'])
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
keep_indexs = (df2['1stFlrSF']<upper)  & (df2['1stFlrSF']>lower) 
df2 = df2[keep_indexs]
train_Y2 = train_Y2[keep_indexs]
```

```python
df2
```

```python
# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df2)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y2, cv=5).mean()
```
